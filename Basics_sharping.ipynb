```python
def reverse_str(x):
    rev = ""
    for i in x:
        rev = i+rev
    return rev
x = "Python"
print(reverse_str(x))
```

```python
def palindrome(word):
    l = 0
    r = len(word)-1
    while l<r:
        if not word[l].isalnum():
            l+=1
        elif not word[r].isalnum():
            r-=1
        elif word[l].lower()!=word[r].lower():
            return False
        else:
            l+=1
            r-=1
    return True

word = "madam"
print(palindrome(word))
```

```python
def freq_char(word):
    count = {}
    for i in word:
        count[i] = count.get(i,0)+1
    high = 0
    freq_char = ""
    for key,val in count.items():
        if val>high:
            high = val
            freq_char = key
    return (freq_char,high)
            
word = "mississippi"
print(freq_char(word))
```

```python
def anagrams(a,b):
    if len(a)!=len(b):
        return False
    count = {}
    for ch in a:
        count[ch] = count.get(ch,0)+1
    for ch in b:
        if ch not in count or count[ch]==0:
            return False
        count[ch]-=1
    return True
     
a = "aab"
b = "abb"
print(anagrams(a,b))
```

```python
def remove_duplicates_maintain_order(word):
    res = ""
    seen = set()
    for ch in word:
        if ch not in seen:
            res+=ch
            seen.add(ch)
    return res
word = "programming"
print(remove_duplicates_maintain_order(word))
```

```python
def grouping_data(data):
    vals = {}
    for item in data:
        group = item['group']
        if group not in vals:
            vals[group] = []
        vals[group].append(item)
    return vals
    
data =[
  {"id": 1, "group": "A"},
  {"id": 2, "group": "B"},
  {"id": 3, "group": "A"},
  {"id": 4, "group": "B"},
  {"id": 5, "group": "C"}
]

print(grouping_data(data))
```

```python
def find_duplicates(data):
    dup =[]
    unq=[]
    for i in data:
        if i not in unq:
            unq.append(i)
        else:
            dup.append(i)
    return dup

data = [1, 2, 3, 2, 4, 5, 1, 6]
print(find_duplicates(data))
```

```python
def find_duplicates(data):
    data.sort()
    dup = []
    l = 0
    for r in range(1,len(data)):
        if data[l] == data[r] and data[l] not in dup:
            dup.append(data[l])
        l+=1
    return dup

data = [1, 2, 3, 2, 4, 5, 1, 6]
print(find_duplicates(data))
```

```python
def sort_by_field(data):
    for i in range(len(data)):
        for j in range(i+1,len(data)):
            if data[i]['age']>data[j]['age']:
                data[i],data[j]=data[j],data[i]
    return data


data = [
  {"name": "John", "age": 25},
  {"name": "Alice", "age": 22},
  {"name": "Bob", "age": 30}
]
print(sort_by_field(data))
```

```python
def sort_by_field(data):
    data.sort(key=lambda x:x['age'])
    return data


data = [
  {"name": "John", "age": 25},
  {"name": "Alice", "age": 22},
  {"name": "Bob", "age": 30}
]
print(sort_by_field(data))
```
