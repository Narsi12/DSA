```python
def reverse_str(x):
    rev = ""
    for i in x:
        rev = i+rev
    return rev
x = "Python"
print(reverse_str(x))
```

```python
def palindrome(word):
    l = 0
    r = len(word)-1
    while l<r:
        if not word[l].isalnum():
            l+=1
        elif not word[r].isalnum():
            r-=1
        elif word[l].lower()!=word[r].lower():
            return False
        else:
            l+=1
            r-=1
    return True

word = "madam"
print(palindrome(word))
```

```python
def freq_char(word):
    count = {}
    for i in word:
        count[i] = count.get(i,0)+1
    high = 0
    freq_char = ""
    for key,val in count.items():
        if val>high:
            high = val
            freq_char = key
    return (freq_char,high)
            
word = "mississippi"
print(freq_char(word))
```

```python
def anagrams(a,b):
    if len(a)!=len(b):
        return False
    count = {}
    for ch in a:
        count[ch] = count.get(ch,0)+1
    for ch in b:
        if ch not in count or count[ch]==0:
            return False
        count[ch]-=1
    return True
     
a = "aab"
b = "abb"
print(anagrams(a,b))
```

```python
def remove_duplicates_maintain_order(word):
    res = ""
    seen = set()
    for ch in word:
        if ch not in seen:
            res+=ch
            seen.add(ch)
    return res
word = "programming"
print(remove_duplicates_maintain_order(word))
```

```python
def threshold_data(data,threshold):
    res = []
    for vals in data:
        age = vals['age']
        if age > threshold:
            res.append(vals)
    return res

data = [
    {"name": "John", "age": 25},
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 20}
]

threshold = 25
print(threshold_data(data,threshold))
```

```python

def flatten(data):
    flat = []
    for item in data:
        if isinstance(item,list):
            flat.extend(flatten(item))
        else:
            flat.append(item)
    return flat

data = [[1, 2], [3, 4], [5]]
print(flatten(data))
```

```python

def second_largest(nums):
    first = second = float('-inf')
    for num in nums:
        if num > first:
            second = first
            first = num
        elif num > second and num!=first:
            second = num
    return second


nums = [10, 20, 4, 45, 99]
print(second_largest(nums))
```

```python

def remove_duplicate_keys(d1,d2):
    c = d1.copy()
    for key,val in d2.items():
        if key in c:
            c[key]+=val
        else:
            c[key]=val
        
    return c
    
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
print(remove_duplicate_keys(d1,d2))
```

```python

def max_key(d):
    maxy = float('-inf')
    common = None
    for key,val in d.items():
        if val > maxy:
            maxy = val
            common = key
    return common
    
d = {'a': 3, 'b': 7, 'c': 2}
print(max_key(d))
```

```python

def fizzBuzz(n):
    new = []
    for i in range(1,n+1):
        if i % 3==0 and i%5== 0:
            new.append("FizzBuzz")
        elif i%3 == 0:
            new.append("Fizz")
        elif i%5==0:
            new.append("Buzz")
        else:
            new.append(i)
    return new
n = 16
print(fizzBuzz(n))
```

```python

def checkPrime(n):
    if n<=1:
        return False
    for num in range(2,n):
        if n%num ==0:
            return False
    return True
        
n = 11
print(checkPrime(n))
```

```python

def fibbo(n):
    fib_series = [0,1]
    for i in range(2,n):
        fib_series.append(fib_series[i-1] + fib_series[i-2])
    return fib_series
        
n = 11
print(fibbo(n))
```

```python

def amstrong(n):
    temp = n
    val = 0
    while n > 0:
        rem = n%10
        n = n//10
        val += rem**3
        
    if val == temp:
        return True
    else:
        return False
        
n = 153
print(amstrong(n))
```

```python
def reverse_number(x):
    y = ''
    while x > 0:
        rem = x%10
        y=y+str(rem)
        x = x//10
    return int(y)

x = 12345
print(reverse_number(x))
```

```python
def sum_of_digits_until_single_digit(x):
    while x>=10:
        y = 0
        while x > 0:
            rem = x%10
            y+=rem
            x = x//10
        x=y
    return x
    

x = 9875
print(sum_of_digits_until_single_digit(x))
```

```python
def sort_employees_sal_des_name_asc(employees):
    n = len(employees)
    for i in range(n):
        for j in range(0,n-i-1):
            if employees[j]['salary']<employees[j+1]['salary']:
                employees[j],employees[j+1]=employees[j+1],employees[j]
            elif employees[j]['salary'] == employees[j+1]['salary']:
                if employees[j]['name']>employees[j+1]['name']:
                    employees[j],employees[j+1]=employees[j+1],employees[j]
    return employees
        
    

employees = [
    {"name": "Ravi", "salary": 50000},
    {"name": "Anil", "salary": 70000},
    {"name": "Kiran", "salary": 50000},
    {"name": "Bhanu", "salary": 70000}
]

print(sort_employees_sal_des_name_asc(employees))
```

```python
def sub_array(arr):
    res = []
    n = len(arr)
    for i in range(n):
        total = 0
        for j in range(i,n):
            total+=arr[j]
            if total == 0:
                res.append(arr[i:j+1])
    return res
        
arr = [1, -1, 2, -2, 3]
print(sub_array(arr))
```
