HOW OPPOSITE ENDS WORKS (Quick Reminder):
  - Place L at index 0, R at last index
  - Move them toward each other based on a condition
  - Stop when L >= R
  - Time O(n), Space O(1)

```python
# LeetCode: https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/
# 1. Two sum II
# INPUT:
#   numbers = [2, 7, 11, 15], target = 9
 
# OUTPUT:
#   [0, 1]   (numbers[0] + numbers[1] = 2 + 7 = 9)

# BruteForce :
 
def two_sum_brute(numbers, target):
      for i in range(len(numbers)):
          for j in range(i + 1, len(numbers)):
              if numbers[i] + numbers[j] == target:
                  return [i , j ]

numbers = [2, 7, 11, 15]
target = 9
print(two_sum_brute(numbers, target))



# Optimal or Better solution
def two_sum_better(nums,target):
    n = len(nums)
    left = 0
    right = n-1
    while left<right:
        total = nums[left]+nums[right]
        if total==target:
            return ([left,right])
        if total < target:
            left+=1
        else:
            right-=1

nums = [2, 7, 11, 15]
target = 9
print(two_sum_better(nums,target))

```

```python
#2 Valid Palindrome
# LeetCode: https://leetcode.com/problems/valid-palindrome/
 
# PROBLEM:
#   A string is a palindrome if it reads the same forward and backward.
#   Consider only alphanumeric characters, ignore cases.
#   Return True if the string is a palindrome.
 
# INPUT:
#   s = "A man, a plan, a canal: Panama"
 
# OUTPUT:
#   True   ("amanaplanacanalpanama" is a palindrome)

# Brute Force
def palindrome(s):
    new=""
    for i in s:
        if i.isalnum():
            new+=i.lower()
    
    if new == new[::-1]:
        return "Palindrome"
    else:
        return "Not a palindrome"
    
s = "A man, a plan, a canal: Panama"
palindrome(s)

# Optimal solution proper optimal way for all cases
def plaindromBetter(s):
    left = 0
    right = len(s)-1
    while left<right:
        if not s[left].isalnum():
            left+=1
        elif not s[right].isalnum():
            right-=1
        elif s[left].lower()!=s[right].lower():
            return "No Plaindrome"
        else:
            left+=1
            right-=1
    return "Plaindrome"

s = "A man, a plan, a canal: Panama"
print(plaindromBetter(s))

# Another question without special characters

def plaindromeOptimal(s):
    s = s.lower()
    start = 0
    end = len(s)-1
    while start<end:
        if s[start]!=s[end]:
            return "No palindrome"
        start+=1
        end-=1
    return "Palindrome"

s="Madam"
print(plaindromeOptimal(s))

# Checking longest palindrome string
def longest_palindrom(s):
    def expand_middle(left,right):
        while left>=0 and right <len(s) and s[left]==s[right]:
            left-=1
            right+=1
        return s[left+1:right]
    longest = ""
    for i in range(len(s)):
        odd = expand_middle(i,i)
        even = expand_middle(i,i+1)
        if len(odd) > len(longest):
            longest = odd
        if len(even) > len(longest):
            longest = even
    return longest

print(longest_palindrom("forgeekskeegfor"))
```

```python
# 3 reverse a string
# LeetCode: https://leetcode.com/problems/reverse-string/
 
# PROBLEM:
#   Reverse a character array IN-PLACE.
#   Must do it with O(1) extra space.
 
# INPUT:
#   s = ['h','e','l','l','o']
 
# OUTPUT:
#   ['o','l','l','e','h']

# def reverse_str(s):
#     a = s
#     return a[::-1]
# s = ['h','e','l','l','o']
# print(reverse_str(s))

#optimal solution
def rev_str(s):
    l = 0
    r = len(s)-1
    while l<r:
        s[l],s[r]=s[r],s[l]
        l+=1
        r-=1
        
    return s

s = ['h','e','l','l','o']
print(reverse_str(s))
```

```python
# Squares of number
# LeetCode: https://leetcode.com/problems/squares-of-a-sorted-array/
 
# PROBLEM:
#   Given a sorted array of integers (may have negatives),
#   return an array of the squares of each number, sorted.
 
# INPUT:
#   nums = [-4, -1, 0, 3, 10]
 
# OUTPUT:
#   [0, 1, 9, 16, 100]
def sqrts(nums):
    for i in range(len(nums)):
        val = nums[i]**2
        nums[i]=val
    return sorted(nums)
    
nums = [-4, -1, 0, 3, 10]
# print(sqrts(nums))

def sq_nums(nums):
    l = 0
    r = len(nums)
    d = []
    while l<r:
        val = nums[l]**2
        nums[l] = val
        l+=1
    return sorted(nums)

nums = [-4, -1, 0, 3, 10]
print(sq_nums(nums))
```

```python
# LeetCode: https://leetcode.com/problems/is-subsequence/
 
# PROBLEM:
#   Given strings s and t, return True if s is a subsequence of t.
#   Subsequence = characters appear in same order but not necessarily adjacent.
 
# INPUT:
#   s = "ace", t = "abcde"
 
# OUTPUT:
#   True   (a...c...e appears in order in "abcde")


s = "ace"
t = "abcde"
l = 0
r = 0
while l<len(s) and r < len(t):
    if s[l] == t[r]:
        l+=1
    r+=1
if len(s) == l:
    print("True")
else:
    print("No")
```

```python
# LeetCode: https://leetcode.com/problems/remove-duplicates-from-sorted-array/
 
# PROBLEM:
#   Given a sorted array, remove duplicates in-place.
#   Return the number of unique elements k.
#   First k elements of the array should hold unique values.
 
# INPUT:
#   nums = [1, 1, 2, 2, 3]
 
# OUTPUT:
#   k = 3, nums = [1, 2, 3, _, _]
def dup_remove(k,nums):
    n = len(nums)
    i = 0
    for j in range(1,n):
        if nums[i]!=nums[j]:
            i+=1
        nums[i] = nums[j] 
    return nums[:k]
k = 3
nums = [1, 1, 2, 2, 3]
print(dup_remove(k,nums))
```

```python
# LeetCode: https://leetcode.com/problems/reverse-vowels-of-a-string/
 
# PROBLEM:
#   Given a string, reverse only the vowels.
#   Return the modified string.
 
# INPUT:
#   s = "hello"
 
# OUTPUT:
#   "holle"   (e and o swapped)

s = "hello"
vowels = "aeiou"
v = []
for i in s:
    if i in vowels:
        v.append(i)
v = v[::-1]
res = []
j = 0
for ch in s:
    if ch in vowels:
        res.append(v[j])
        j+=1
    else:
        res.append(ch)
print("".join(res))
        
```

```python
def vowels_swap(s):
    n = len(s)
    s = list(s)
    vowels = set("AEIOUaeiou")
    l = 0
    r = n-1
    while l<r:
        while l<r and s[l] not in vowels:
            l+=1
        while l<r and s[r] not in vowels:
            r-=1
        if l<r:
            s[l],s[r]=s[r],s[l]
            l+=1
            r-=1
    return "".join(s)

s = "hello"
print(vowels_swap(s))
```

```python
# LeetCode: https://leetcode.com/problems/4sum/
 
# PROBLEM:
#   Find all unique quadruplets that sum to target.
 
# INPUT:
#   nums = [1, 0, -1, 0, -2, 2], target = 0
 
# OUTPUT:
#   [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]

def four_sum(nums,target):
    n = len(nums)
    x = []
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                for l in range(k+1,n):
                    total = nums[i]+nums[j]+nums[k]+nums[l]
                    if total == target:
                        x.append(tuple(sorted([nums[i],nums[j],nums[k],nums[l]])))
    return x

nums = [1, 0, -1, 0, -2, 2]
target = 0
print(four_sum(nums,target))
```

```python
def sorted_nums_val(nums):
    left = 0
    while left < len(nums)-1 and nums[left]<=nums[left+1]:
        left+=1
    if left == len(nums)-1:
        return 0
    right = len(nums)-1
    while right>0 and nums[right]>=nums[right-1]:
        right-=1
    sub = sorted(nums[left:right+1])
    nums[left:right+1] = sub
    return nums

nums = [2, 6, 4, 8, 10, 9, 15]
# nums = [1,2,3,4,5,6]
sorted_nums_val(nums)
```

```python
def three_sum(nums):
    nums.sort()
    n = len(nums)
    res = []
    for i in range(n):
        if i > 0 and nums[i]==nums[i-1]:
            continue
        left=i+1
        right=n-1
        while left<right:
            total = nums[i]+nums[left]+nums[right]
            if total < 0:
                left+=1
            elif total > 0:
                right-=1
            else:
                res.append([nums[i],nums[left],nums[right]])
                while left<right and nums[left]==nums[left+1]:
                    left+=1
                while left<right and nums[right]==nums[right-1]:
                    right-=1
            left+=1
            right-=1
    return res
            

nums = [-1,0,1,2,-1,-4]
print(three_sum(nums))
```

```python
# container with most water
def maxArea(height):
    max_area = 0
    n = len(height)
    l = 0
    r = n-1
    while l<r:
        w = r-l
        h = min(height[l],height[r])
        area = h*w
        max_area = max(area,max_area)
        if height[l]<=height[r]:
            l+=1
        else:
            r-=1
    return max_area
height = [1,8,6,2,5,4,8,3,7]
print(maxArea(height))
```

```python
# Trapping Rain water
def tarpping_rainwater(height):
    n = len(height)
    total_water = 0
    for i in range(n):
        left_max = 0
        right_max = 0
        for j in range(i+1):
            left_max = max(left_max,height[j])
        for j in range(i,n):
            right_max = max(right_max,height[j])
        total_water+= min(left_max,right_max)-height[i]
    return total_water
        
    
height = [0,1,0,2,1,0,1,3,2,1,2,1]
# height =  [3, 0, 2, 4]
print(tarpping_rainwater(height))
```
