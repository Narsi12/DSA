# Stack

A Stack is a linear data structure that follows:

LIFO → Last In First Out

The element inserted last will be removed first.

# Real-Time Example
Think about a stack of plates.
Plate 1
Plate 2
Plate 3
If you want to remove a plate, you always remove the top plate first.
So:
Last plate added → removed first

# Stack Operations
1. Push
Adds element to top of stack.
ex:
stack.append(10)

3. Pop
Removes top element.
ex:
stack.pop()

3. Peek / Top
Gets top element without removing.
ex:
stack[-1]

# Stack Visualization

# Why Stack is Used?
Use Cases
->Function calls
->Undo operation
->Browser back button
->Expression evaluation
->Recursion handling

Stack in Python
->Python list acts like a stack because:
append() → insert at end
pop() → remove from end
Both operations are fast.

# Queue
We have to import the queue from python collections package

from collections import deque

->A Queue is a linear data structure that follows:
FIFO → First In First Out
The element inserted first will be removed first.

Real-Time Example
->Think about a ticket counter line.
Person A
Person B
Person C
Person A entered first, so Person A leaves first.

Queue Operations
1. Enqueue
->Insert element at rear/end.
Example:
queue.append(10)

2. Dequeue
->Remove element from front.
Example:
queue.popleft()

3. Front
->Get first element.
Example:
queue[0]

# Queue Visualization
Insert:
10
20
30
Queue becomes:
Front → 10 20 30 ← Rear
Now dequeue:
10 removed
Remaining:
20 30

# Why Queue is Used?
Use Cases
->Task scheduling
->Printer queue
->CPU scheduling
->Message queues
->BFS traversal

Why Use deque Instead of List?
Removing first element from list:
pop(0)
is slow because elements shift.

But deque is optimized for:
append()
popleft()
So Queue implementation is faster using:
collections.deque

```python
def is_valid(s):
    mapping = {'}':'{',']':'[',')':'('}
    stack = []
    for item in s:
        if item in '([{':
            stack.append(item)
        else:
            top_ele = stack.pop() if stack else ""
            if top_ele != mapping[item]:
                return False
         
    return not stack
        

s = "()[]{}"
print(is_valid(s))
```

```python
def reverse_string(s):
    stack = []
    for i in s:
        stack.append(i)
    x = reversed(stack)
    return "".join(x)


s = "python"
print(reverse_string(s))
```

```python
def remove_duplicates(s):
    stack = []
    for i in s:
        if not stack or stack[-1]!=i:
            stack.append(i)
    return "".join(stack)


s = "aabbccdde"
print(remove_duplicates(s))
```

```python
def remove_duplicates(s):
    stack = []
    for i in s:
        if stack and stack[-1]==i:
            stack.pop()
        else:
            stack.append(i)
    return "".join(stack)


s = "aabbccdde"
print(remove_duplicates(s))
```

```python
def evaluate_postfix(expression):
    stack = []
    for i in expression:
        if i.isdigit():
            stack.append(int(i))
        else:
            if len(stack)<2:
                return "Nope"
            a = stack.pop()
            b = stack.pop()
            if i == "+":
                stack.append(a+b)
            elif i == "*":
                stack.append(a*b)
            elif i == "-":
                stack.append(a-b)
            elif i == "/":
                stack.append(a//b)
    return stack

expression = "23*5+"

print(evaluate_postfix(expression))
```

```python
from collections import deque
def queue_methods():
    q = deque()
    q.append(10)
    q.append(20)
    print(q[0])
    print(q.popleft())  # FIFO
    print(q.popleft())  
    print(len(q) == 0) 
    return q
print(queue_methods())
```

```python
from collections import deque

def reverse_k(q,k):
    stack = []
    for _ in range(k):
        stack.append(q.popleft())
    while stack:
        q.append(stack.pop()) # here the pop will remove last values 3,2,1 and append in 'q' as 3,2,1
        
    for _ in range(len(q)-k): # 5-3 = 2 so here we take 4,5 popleft()
        q.append(q.popleft())
        
    return q


q = deque([1, 2, 3, 4, 5])
k = 3

print(reverse_k(q, k))
```

```python
from collections import deque

def generate_binary(n):
    q = deque()
    q.append('1')
    for _ in range(n):
        font = q.popleft()
        print(font,end=" ")
        q.append(font+"0")
        q.append(font+"1")
    

n = 5
print(generate_binary(n))
```
