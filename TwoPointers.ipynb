HOW OPPOSITE ENDS WORKS (Quick Reminder):
  - Place L at index 0, R at last index
  - Move them toward each other based on a condition
  - Stop when L >= R
  - Time O(n), Space O(1)

```python
# LeetCode: https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/
# 1. Two sum II
# INPUT:
#   numbers = [2, 7, 11, 15], target = 9
 
# OUTPUT:
#   [0, 1]   (numbers[0] + numbers[1] = 2 + 7 = 9)

# BruteForce :
 
def two_sum_brute(numbers, target):
      for i in range(len(numbers)):
          for j in range(i + 1, len(numbers)):
              if numbers[i] + numbers[j] == target:
                  return [i , j ]

numbers = [2, 7, 11, 15]
target = 9
print(two_sum_brute(numbers, target))



# Optimal or Better solution
def two_sum_better(nums,target):
    n = len(nums)
    left = 0
    right = n-1
    while left<right:
        total = nums[left]+nums[right]
        if total==target:
            return ([left,right])
        if total < target:
            left+=1
        else:
            right-=1

nums = [2, 7, 11, 15]
target = 9
print(two_sum_better(nums,target))

```

```python
#2 Valid Palindrome
# LeetCode: https://leetcode.com/problems/valid-palindrome/
 
# PROBLEM:
#   A string is a palindrome if it reads the same forward and backward.
#   Consider only alphanumeric characters, ignore cases.
#   Return True if the string is a palindrome.
 
# INPUT:
#   s = "A man, a plan, a canal: Panama"
 
# OUTPUT:
#   True   ("amanaplanacanalpanama" is a palindrome)

# Brute Force
def palindrome(s):
    new=""
    for i in s:
        if i.isalnum():
            new+=i.lower()
    
    if new == new[::-1]:
        return "Palindrome"
    else:
        return "Not a palindrome"
    
s = "A man, a plan, a canal: Panama"
palindrome(s)

# Optimal solution proper optimal way for all cases
def plaindromBetter(s):
    left = 0
    right = len(s)-1
    while left<right:
        if not s[left].isalnum():
            left+=1
        elif not s[right].isalnum():
            right-=1
        elif s[left].lower()!=s[right].lower():
            return "No Plaindrome"
        else:
            left+=1
            right-=1
    return "Plaindrome"

s = "A man, a plan, a canal: Panama"
print(plaindromBetter(s))

# Another question without special characters

def plaindromeOptimal(s):
    s = s.lower()
    start = 0
    end = len(s)-1
    while start<end:
        if s[start]!=s[end]:
            return "No palindrome"
        start+=1
        end-=1
    return "Palindrome"

s="Madam"
print(plaindromeOptimal(s))

# Checking longest palindrome string
def longest_palindrom(s):
    def expand_middle(left,right):
        while left>=0 and right <len(s) and s[left]==s[right]:
            left-=1
            right+=1
        return s[left+1:right]
    longest = ""
    for i in range(len(s)):
        odd = expand_middle(i,i)
        even = expand_middle(i,i+1)
        if len(odd) > len(longest):
            longest = odd
        if len(even) > len(longest):
            longest = even
    return longest

print(longest_palindrom("forgeekskeegfor"))
```

```python
# 3 reverse a string
# LeetCode: https://leetcode.com/problems/reverse-string/
 
# PROBLEM:
#   Reverse a character array IN-PLACE.
#   Must do it with O(1) extra space.
 
# INPUT:
#   s = ['h','e','l','l','o']
 
# OUTPUT:
#   ['o','l','l','e','h']

# def reverse_str(s):
#     a = s
#     return a[::-1]
# s = ['h','e','l','l','o']
# print(reverse_str(s))

#optimal solution
def rev_str(s):
    l = 0
    r = len(s)-1
    while l<r:
        s[l],s[r]=s[r],s[l]
        l+=1
        r-=1
        
    return s

s = ['h','e','l','l','o']
print(reverse_str(s))
```

```python
def three_sum(nums):
    nums.sort()
    n = len(nums)
    res = []
    for i in range(n):
        if i > 0 and nums[i]==nums[i-1]:
            continue
        left=i+1
        right=n-1
        while left<right:
            total = nums[i]+nums[left]+nums[right]
            if total < 0:
                left+=1
            elif total > 0:
                right-=1
            else:
                res.append([nums[i],nums[left],nums[right]])
                while left<right and nums[left]==nums[left+1]:
                    left+=1
                while left<right and nums[right]==nums[right-1]:
                    right-=1
            left+=1
            right-=1
    return res
            

nums = [-1,0,1,2,-1,-4]
print(three_sum(nums))
```

```python
# container with most water
def maxArea(height):
    max_area = 0
    n = len(height)
    l = 0
    r = n-1
    while l<r:
        w = r-l
        h = min(height[l],height[r])
        area = h*w
        max_area = max(area,max_area)
        if height[l]<=height[r]:
            l+=1
        else:
            r-=1
    return max_area
height = [1,8,6,2,5,4,8,3,7]
print(maxArea(height))
```

```python
# Trapping Rain water
def tarpping_rainwater(height):
    n = len(height)
    total_water = 0
    for i in range(n):
        left_max = 0
        right_max = 0
        for j in range(i+1):
            left_max = max(left_max,height[j])
        for j in range(i,n):
            right_max = max(right_max,height[j])
        total_water+= min(left_max,right_max)-height[i]
    return total_water
        
    
# height = [0,1,0,2,1,0,1,3,2,1,2,1]
height =  [3, 0, 2, 4]
print(tarpping_rainwater(height))
```
