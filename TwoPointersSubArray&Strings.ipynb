```python
def sub_array_sum_of_k(nums,k):
    window = 0
    n = len(nums)
    freq = {}
    for i in range(k):
        window += nums[i]
        freq[nums[i]] = freq.get(nums[i],0)+1
        
    max_sum = window if len(freq) ==k else 0
    for j in range(k,n):
        window += nums[j]
        freq[nums[j]] = freq.get(nums[j],0)+1
        window-=nums[j-k]
        freq[nums[j-k]]-=1
        if freq[nums[j-k]]==0:
            del freq[nums[j-k]]
        if len(freq) == k:
            max_sum = max(max_sum,window)
    return max_sum
        
nums = [1,5,4,2,9,9,9]
k = 3
print(sub_array_sum_of_k(nums,k))
```
